```
regrelinear = pd.read_csv("comp_bikes_mod.csv")
```

```
regrelinear.keys()
```

```
#Poderia ter utilizado o numpy.column_stack(())
regrelinear["hum"].fillna(regrelinear["hum"].mean(), inplace=True )
regrelinear["casual"].fillna(regrelinear["casual"].mean(), inplace=True)
regrelinear["cnt"].fillna(regrelinear["cnt"].mean(),inplace=True)
```

```
regrelinear.isnull().sum()
import numpy as np
```

```
#Utilizando o Reshape para criar um array com duas colunas
X_regrelinear_casual = regrelinear.iloc[:,-3].values
X_regrelinear_hum = regrelinear.iloc[:,-5].values
Y_regrelinear = regrelinear.iloc[:,-1].values

X_regrelinear = np.concatenate((X_regrelinear_casual.reshape(-1,1),X_regrelinear_hum.reshape(-1,1)), axis=1)

```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error 
```

```
#construindo o modelo de regressão Linear 
reg = LinearRegression()
regressao = reg.fit(X_regrelinear,Y_regrelinear)
```

```
#A fazer a construção da previsão para TREINO
previsao = regressao.predict(X_regrelinear)
previsao
```

```
#A calcular o valor do erro médio quadrado
MSE = mean_squared_error(Y_regrelinear,previsao) 
MSE
```

```
#Como ficaria o cálculo

print( "Y = {}X {} ".format(reg.coef_,reg.intercept_))
from sklearn.metrics import r2_score

R2 = r2_score(Y_regrelinear,previsao)
print("Coeficiente de determinação é {} .".format(R2))


```
